### 3. Variable Visualization [3 pts]

#### 3.1. Create a barplot.

Select a categorical variable from this dataset and create a barplot that shows the *counts* for each level of this categorical variable.

```python
category_count = df['category'].value_counts()
display(category_count)
```

```python
df['category']
```

```python
sns.barplot(x = category_count.index, y = category_count)
plt.title('Number of teams in each category')
plt.xlabel('category')
plt.ylabel('number of teams')

plt.show()
```

#### We found that the labels on x-axis squeeze together and is hard to read.
#### One way to solve it is to use set_xticklabels() command to set the rotation and fontsize of labels

```python
fig = sns.barplot(x = category_count.index, y = category_count)

fig.set_xticklabels(labels = category_count.index, rotation = 90) # This is what we did

plt.title('Which category plays more games?')
plt.xlabel('category')
plt.ylabel('number of teams')

plt.show()
```
